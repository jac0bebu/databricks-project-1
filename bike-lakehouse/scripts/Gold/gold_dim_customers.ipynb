#Transformation Process

```python
df = spark.sql("""
    SELECT 
        c.customer_id AS customer_key, 
        c.customer_number, 
        c.first_name,
        c.last_name,
        c.marital_status,
        c.gender,
        cl.country,
        c_erp.birth_date,
        c.created_date
    FROM workspace.silver.crm_customers c 
    LEFT JOIN workspace.silver.erp_customer_location cl 
        ON c.customer_number = cl.customer_number
    LEFT JOIN workspace.silver.erp_customers c_erp 
        On c.customer_number = c_erp.customer_number
                    """)
```

```python
df.limit(19).display()
```

#Writing into Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_customers")
```
