#Transformation Process

```python
df = spark.sql("""
    SELECT 
        p.product_id as product_key,
        p.product_number,
        p.product_name,
        p.product_cost,
        pc.category_id,
        pc.category,
        pc.subcategory,
        pc.maintenance_flag,
        p.product_line,
        p.product_start_date,
        p.product_end_date
    FROM workspace.silver.crm_products p
    LEFT JOIN workspace.silver.erp_product_category pc
        ON p.product_catagory = pc.category_id
                    """)
```

```python
df.limit(19).display()
```

#Writing into Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
